```python
import pandas as pd

# 1️⃣ Read CSV
data = pd.read_csv("Muhurta_2027_2.csv", encoding="latin1")
data.isna().sum()
# 2️⃣ Convert Date to datetime FIRST
data["Date"] = pd.to_datetime(
    data["Date"],
    errors="coerce",
    dayfirst=True
)

# 3️⃣ Sort while it's datetime
data = data.sort_values("Date")

# 4️⃣ Clean text columns (non-date)
def clean_text(value):
    if isinstance(value, str):
        return (
            value.replace("\u0096", "-")
                 .replace("\u2013", "-")
                 .replace("\u2014", "-")
        )
    return value

print(type(data))

data = data.applymap(clean_text)

# 5️⃣ Format Date ONCE (calendar format)
data["Date"] = data["Date"].dt.strftime("%Y-%m-%d")
```

```python
df = pd.read_csv('Muhurta_2027_2.csv', encoding="latin1")

# Clean date column first
df["Date"] = (
    df["Date"]
    .astype(str)
    .str.strip()
)

df["Date"] = pd.to_datetime(
    df["Date"],
    format="%m/%d/%Y",   # since your CSV is MM/DD/YYYY
    errors="coerce"
)

df = df.dropna(subset=["Date"])

# Clean text columns safely
def clean_text(value):
    if isinstance(value, str):
        return (
            value.replace("\u0096", "-")
                 .replace("\u2013", "-")
                 .replace("\u2014", "-")
        )
    return value

for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].map(clean_text)

df["Date"] = df["Date"].dt.strftime("%Y-%m-%d")
```
